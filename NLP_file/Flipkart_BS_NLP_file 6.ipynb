```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
```

# Flipkart.com

```python
review_list = []
```

```python
def get_soup(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.text,"html.parser")
    return soup
```

```python
def get_reviews(soup):
    reviews=soup.find_all('div',{'class' : '_27M-vq'})
    try:
        for item in reviews:
            review={    
            'review' : item.find('div',{'class':'t-ZTKy'}).text.strip(),
            'rating': float(item.find('div',{'class':'_3LWZlK'}).text.strip())
            }
            review_list.append(review)
            
    except:
        pass
```

```python
for i in range(1,1200):
    soup = get_soup(f'https://www.flipkart.com/mi-smart-band-4/product-reviews/itmff9f94c0d486e?pid=SBNFPNVT2FJEKKHV&lid=LSTSBNFPNVT2FJEKKHVYORHRM&marketplace=FLIPKART&page={i}')
    print(f'Getting Page: {i}')
    get_reviews(soup)
    print(len(review_list))
```

```python
df = pd.DataFrame(review_list)
df.to_csv('Ratings_nlp_flip_6.csv', index=False)
print('END')
```
