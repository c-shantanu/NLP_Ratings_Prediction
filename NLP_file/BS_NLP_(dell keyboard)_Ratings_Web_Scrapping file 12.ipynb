```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
```

# amazon.in

```python
review_list = []
```

```python
def get_soup(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.text,"html.parser")
    return soup
```

```python
def get_reviews(soup):
    reviews=soup.find_all('div',{'data-hook' : 'review'})
    try:
        for item in reviews:
            review={    
            'review' : item.find('span',{'data-hook':'review-body'}).text.strip(),
            'rating': float(item.find('i',{'data-hook':'review-star-rating'}).text.replace('out of 5 stars',"").strip())
            }
            review_list.append(review)
    except:
        pass
```

```python
for i in range(1,999):
    soup = get_soup(f'https://www.amazon.in/Dell-Km117-Wireless-Keyboard-Mouse/product-reviews/B01LOORNLY/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber={i}')
    print(f'Getting Page: {i}')
    get_reviews(soup)
    print(len(review_list))
    if not soup.find('li', {'class':'a-disabled a-last'}):
        pass
    else:
        break
```

```python
df = pd.DataFrame(review_list)
df.to_csv('Ratings_nlp_12.csv', index=False)
print('END')
```
